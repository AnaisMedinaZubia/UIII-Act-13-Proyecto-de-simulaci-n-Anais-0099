#### Problema: credito hipoticario

```python
# --- SISTEMA DE IA: ASISTENTE HIPOTECARIO INTELIGENTE ---

# Lista global para almacenar los objetos de crédito [6]
base_datos_creditos = []

class Credito_hipotecario:
    """Clase que representa un crédito de vivienda con lógica de IA [1, 7]"""

    def __init__(self, numero, monto, tasa, plazo):
        # El método __init__ inicializa los atributos del objeto [8, 9]
        self.numero_credito = numero
        self.monto_total_prestado = monto
        self.saldo_pendiente = monto
        self.tasa_interes_anual = tasa
        self.plazo_restante_meses = plazo

    def realizar_pago_mensual(self, monto_pago):
        """Resta el pago al saldo pendiente utilizando operadores aritméticos [10, 11]"""
        if monto_pago > 0:
            self.saldo_pendiente -= monto_pago
            self.plazo_restante_meses -= 1
            print(f"\n[SISTEMA] Pago de ${monto_pago} aplicado con éxito.")
        else:
            print("\n[ERROR] El pago debe ser mayor a cero.")

    def consultar_estado_cuenta(self):
        """Simula un diagnóstico de IA sobre la salud del crédito [18.1, 1002]"""
        # Lógica de decisión basada en reglas (IA básica) [12, 13]
        progreso = (self.monto_total_prestado - self.saldo_pendiente) / self.monto_total_prestado

        if progreso >= 0.50:
            ia_mensaje = "EXCELENTE: Has pagado más de la mitad. Tu salud financiera es sólida."
        elif self.saldo_pendiente > self.monto_total_prestado * 0.90:
            ia_mensaje = "ALERTA: El saldo es muy alto. Considera realizar pagos anticipados."
        else:
            ia_mensaje = "ESTABLE: Tu crédito avanza según lo planeado."

        return ia_mensaje

    def aplicar_interes_mensual(self):
        """Calcula e incrementa el interés mensual al saldo [14, 15]"""
        interes_mensual = self.saldo_pendiente * (self.tasa_interes_anual / 100 / 12)
        self.saldo_pendiente += interes_mensual
        print(f"\n[IA] Interés mensual aplicado: ${interes_mensual:.2f}")

    def simular_pago_anticipado(self, abono_capital):
        """La IA proyecta el ahorro si se paga por adelantado [16]"""
        ahorro_estimado = abono_capital * (self.tasa_interes_anual / 100) * (self.plazo_restante_meses / 12)
        print(f"\n--- PROYECCIÓN IA DE PAGO ANTICIPADO ---")
        print(f"Si abonas ${abono_capital} hoy, ahorrarías aproximadamente ${ahorro_estimado:.2f} en intereses futuros.")

    def modificar_plazo_por_reestructura(self, nuevo_plazo):
        """Actualiza el plazo restante [17]"""
        self.plazo_restante_meses = nuevo_plazo
        print(f"\n[SISTEMA] Reestructuración exitosa. Nuevo plazo: {nuevo_plazo} meses.")

# --- FUNCIONES DE INTERFAZ DEL MENÚ ---

def agregar_credito():
    print("\n--- REGISTRO DE NUEVO CRÉDITO HIPOTECARIO ---")
    numero = input("Número de crédito: ")
    monto = float(input("Monto total prestado: "))
    tasa = float(input("Tasa de interés anual (%): "))
    plazo = int(input("Plazo restante (meses): "))

    # Instanciación: Creación física del objeto en memoria [18-20]
    nuevo_c = Credito_hipotecario(numero, monto, tasa, plazo)
    base_datos_creditos.append(nuevo_c)
    print(f"Crédito {numero} registrado.")

def mostrar_tabla():
    if len(base_datos_creditos) == 0:
        print("\n[!] No hay créditos registrados.")
    else:
        # Encabezado alineado con f-strings para una buena distribución [21, 22]
        header = f"{'NÚMERO':<10} | {'MONTO TOTAL':<15} | {'SALDO':<15} | {'TASA %':<8} | {'MESES':<8}"
        print("\n" + "="*70)
        print(header)
        print("-" * 70)

        # Ciclo for para recorrer la lista de objetos [3, 23]
        for c in base_datos_creditos:
            print(f"{c.numero_credito:<10} | ${c.monto_total_prestado:<14.2f} | ${c.saldo_pendiente:<14.2f} | {c.tasa_interes_anual:<8.2f} | {c.plazo_restante_meses:<8}")
        print("="*70)

# --- MENÚ PRINCIPAL INTERACTIVO [24, 25] ---
while True:
    print("\n========================================")
    print("   CENTRO FINANCIERO INTELIGENTE (IA)   ")
    print("========================================")
    print("1. Registrar Crédito")
    print("2. Listado de Créditos (Tabla)")
    print("3. Realizar Pago Mensual")
    print("4. IA: Consultar Diagnóstico de Salud")
    print("5. IA: Simular Pago Anticipado")
    print("6. Salir")

    opcion = input("\nSeleccione una opción: ")

    if opcion == "1":
        agregar_credito()
    elif opcion == "2":
        mostrar_tabla()
    elif opcion == "3":
        num = input("Número de crédito: ")
        encontrado = False
        for c in base_datos_creditos:
            if c.numero_credito == num:
                pago = float(input("Monto a pagar: "))
                c.aplicar_interes_mensual()
                c.realizar_pago_mensual(pago)
                encontrado = True
        if not encontrado: print("Crédito no encontrado.")
    elif opcion == "4":
        num = input("Número de crédito para diagnóstico: ")
        for c in base_datos_creditos:
            if c.numero_credito == num:
                print(f"\n[IA DIAGNÓSTICO]: {c.consultar_estado_cuenta()}")
    elif opcion == "5":
        num = input("Número de crédito: ")
        for c in base_datos_creditos:
            if c.numero_credito == num:
                abono = float(input("Monto de abono extra: "))
                c.simular_pago_anticipado(abono)
    elif opcion == "6":
        print("\nSaliendo del sistema inteligente... ¡Hasta pronto!")
        print("programacion realizado por Anais Medina NL = 40")
        break
    else:
        print("\nOpción no válida.")
```
